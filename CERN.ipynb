```python
import uproot
```

```python
import matplotlib.pyplot as plt
```

```python
fileL = uproot.open("hard_process_L_10k.root:events")
```

```python
fileR = uproot.open("hard_process_R_10k.root:events")
```

```python
fileL.keys()
```

```python
EnergyL=fileL["energy"].array()
```

```python
EnergyR=fileR["energy"].array()
```

```python
MassR=fileR["mass"].array()
```

```python
MassL=fileL["mass"].array()
```

```python
E_tot = []
for Energy in EnergyL:
    E_tot.append(sum(Energy))

plt.hist(E_tot, bins=30,range=(1500,15000))
```

```python
EParticleL_tot = []
for Event in EnergyL:
    for Energy in Event:
        EParticleL_tot.append(Energy)
```

```python
EParticleR_tot=[]
for Event in EnergyR:
    for Energy in Event:
        EParticleR_tot.append(Energy)
```

```python
x=0
y=3500
```

```python
plt.hist(EParticleR_tot, bins=400, range=(x,y), label="RightSamp", histtype="step", color="r")
plt.hist(EParticleL_tot, bins=400, range=(x,y), label="LeftSamp", histtype="bar", color="b", )
plt.xlabel('Energy-Axis (GeV)')
plt.ylabel('Number of Hits-Axis')
plt.title('#Hits Per Energy', fontweight='bold')
plt.legend()
plt.show()
```

```python
MassPL_tot=[]
for Event in MassL:
    for Mass in Event:
        MassPL_tot.append(Mass)
```

```python
MassPR_tot=[]
for Event in MassR:
    for Mass in Event:
        MassPR_tot.append(Mass)
```

```python
z=170
c=180
```

```python
plt.hist(MassPL_tot, bins=200, color="b", histtype="bar", label="Left Hit", range=(z,c))
plt.hist(MassPR_tot, bins=200, color="r", histtype="step", label="Right Hit", range=(z,c))
plt.xlabel('Mass(GeV)')
plt.ylabel('#Hit')
plt.title('Top Quark Mass')
plt.legend()
plt.show
```

```python
f=70
d=90
```

```python
plt.hist(MassPL_tot, bins=200, color="b", histtype="bar", label="Left Hit", range=(f,d))
plt.hist(MassPR_tot, bins=200, color="r", histtype="step", label="Right Hit", range=(f,d))
plt.xlabel('Mass(GeV)')
plt.ylabel('#Hit')
plt.title('W Boson Mass')
plt.legend()
plt.show
```

```python
plt.hist(MassPL_tot, bins=200, color="b", histtype="bar", label="Left Hit", range=(0,200))
plt.hist(MassPR_tot, bins=200, color="r", histtype="step", label="Right Hit", range=(0,200))
plt.xlabel('Mass(GeV)')
plt.ylabel('#Hit')
plt.title('#Hit Per Mass')
plt.legend()
plt.show
```
